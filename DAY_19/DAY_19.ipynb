# matplotlip

<h3>1:Line chart</h3>

```python
import matplotlib.pyplot as plt
```

```python

# Sample data for line chart
x_line = [1, 2, 3, 4, 5]
y_line = [2, 3, 5, 7, 11]

# Create line chart
plt.plot(x_line, y_line, marker='o')
plt.title("Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

<h3>2:Bar chart</h3>

```python
# Sample data for bar chart
categories_bar = ['A', 'B', 'C', 'D']
values_bar = [10, 25, 15, 30]

# Create bar chart
plt.bar(categories_bar, values_bar)
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

<h3>3:Column chart</h3>

```python
categories_bar = ['A', 'B', 'C', 'D']
values_bar = [10, 25, 15, 30]
plt.barh(categories_bar,values_bar)
```

<h3>4:Pie Chart</h3>

```python
# Sample data for pie chart
sizes_pie = [15, 30, 45, 10]
labels_pie = ['Group A', 'Group B', 'Group C', 'Group D']

# Create pie chart
plt.pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

<h3>5:Scatter Plot</h3>

```python
# Sample data for scatter plot
x_scatter = [5, 7, 8, 6, 2, 4, 9, 3]
y_scatter = [10, 14, 12, 13, 7, 8, 15, 9]

# Create scatter plot
plt.scatter(x_scatter, y_scatter)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

<h3>6:Grouped Bar Chart</h3>

```python
import numpy as np

# Sample data for grouped bar chart
categories_grouped = ['A', 'B', 'C']
values1 = [10, 15, 20]
values2 = [5, 10, 15]
bar_width = 0.35
index = np.arange(len(categories_grouped))

# Create grouped bar chart
plt.bar(index, values1, bar_width, label='Group 1')
plt.bar(index + bar_width, values2, bar_width, label='Group 2')
plt.title("Grouped Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.xticks(index + bar_width / 2, categories_grouped)
plt.legend()
plt.show()
```

<h3>7:Box Plot</h3>

```python
# Sample data for box plot
data_box = [12, 15, 14, 10, 13, 18, 20, 22, 19, 17]

# Create box plot
plt.boxplot(data_box)
plt.title("Box Plot")
plt.ylabel("Values")
plt.show()
```

<h3>8:Histogram</h3>

```python
# Sample data for histogram
data_hist = [1, 2, 2, 3, 3, 3, 4, 5, 5, 6, 6, 6, 7, 8, 9, 10]

# Create histogram
plt.hist(data_hist, bins=10)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

<h3>9:Area Chart</h3>

```python
# Sample data for area chart
x_area = [1, 2, 3, 4, 5]
y_area = [1, 3, 2, 5, 4]

# Create area chart
plt.fill_between(x_area, y_area)
plt.title("Area Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

<h3>10:Multiple Line Chart</h3>

```python
# Sample data for multiple line chart
x_multi = [1, 2, 3, 4, 5]
y1_multi = [1, 2, 3, 4, 5]
y2_multi = [5, 4, 3, 2, 1]

# Create multiple line chart
plt.plot(x_multi, y1_multi, label='Series 1')
plt.plot(x_multi, y2_multi, label='Series 2')
plt.title("Multiple Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
